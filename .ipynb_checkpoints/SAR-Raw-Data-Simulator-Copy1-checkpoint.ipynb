# From the Book#

*LFM pulsed signal*

A0 = 1 amplititude of the transimtter signal

Tp -- pulse duration

fc -- carrier frequency

alpha7s -- LFM pulse chirp rate (range FM rate Î± = 4e12 Hz/s)


```python
# radar beam carrier frequency (Hz)
fc = 1.0e10 
# light speed (speed of wave) (m/s)
c = 3e8
# wavelength of carrier = c/fc (m)
lambd = 0.03
# amplitude of transimtter signal
A0 = 1.0
# time for how long we observe one pulse
time7s = 250
# pulse repetition frequency (Hz                                 )
PRF = 500
# pulse duration (s)
Tp = 6.033e-6
# total pulses 
tot_pulses = PRF * time7s
# LFM pulse chirp rate (Range FM rate) (Hz/s)
alpha7s = 4.0e12
# number of time samples within the LFM pulse (fs*Tp)
tot_samples = 181
```

```python
def generatePulsedLFM(A0, Tp, fc, alpha7s, tot_samples):
    t = np.linspace(0, Tp, tot_samples)
    p = (A0 * np.exp(1j*2*np.pi*fc*t + 1j*np.pi*alpha7s*t*t)).astype(np.complex64)
    return t, p
```

```python
t, p = generatePulsedLFM(A0, Tp, fc, alpha7s, tot_samples)
```

```python
def plot_LFM_pulses(p, tot_samples):
    x = range(0, tot_samples)
    print (len(x))
    y1 = p.real
    y2 = p.imag
    # Two subplots, the axes array is 1-d
    f, axarr = plt.subplots(2, sharex=True)
    axarr[0].plot(x, y1)
    axarr[0].set_title('LFM pulse real part')
    axarr[1].set_title('LFM pulse imaginary part')
    axarr[1].plot(x, y2)
    axarr[1].set_xlabel('time samples')
    axarr[0].set_ylabel('amplitude')
    axarr[1].set_ylabel('amplitude')
```

```python
plot_LFM_pulses(p, tot_samples)
```

```python
def generateIdealChirp(chirpStartFreq,chirpStopFreq,chirpLength,rangeSamplingRate):
    pulseDuration=chirpLength/rangeSamplingRate
    print(pulseDuration);
    k=(chirpStopFreq-chirpStartFreq)/pulseDuration
    t=np.linspace(0,pulseDuration,chirpLength)
    print(t);
    return np.asfortranarray(np.exp(1j*np.pi*(2*chirpStartFreq*t+k*t*t)).reshape(chirpLength,1)).astype(np.complex64)
```

```python
generateIdealChirp(0.01,1,1000,75);
```

```python
def plot_LFM_pulses2(p, tot_samples):
    x = range(0, tot_samples)
    print (len(x))
    y1 = p.real
    y2 = p.imag
    # Two subplots, the axes array is 1-d
    f, axarr = plt.subplots(2, sharex=True)
    axarr[0].plot(x, y1)
    axarr[0].set_title('LFM pulse real part')
    axarr[1].set_title('LFM pulse imaginary part')
    axarr[1].plot(x, y2)
    axarr[1].set_xlabel('time samples')
    axarr[0].set_ylabel('amplitude')
    axarr[1].set_ylabel('amplitude')
```

```python
ideal_chirp = generateIdealChirp(chirpStartFreq=fc,chirpStopFreq=1,chirpLength=Tp,rangeSamplingRate=30e6)
plot_LFM_pulses2(ideal_chirp, len(ideal_chirp))
```
